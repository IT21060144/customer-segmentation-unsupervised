# Unsupervised Learning Project: Customer Segmentation

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, DBSCAN
import matplotlib.pyplot as plt
import seaborn as sns
```

## Simulated Dataset Loading

```python
np.random.seed(42)
data = pd.DataFrame({
    "Annual Income (k$)": np.random.normal(60, 20, 200).clip(15, 150),
    "Spending Score (1-100)": np.random.randint(1, 101, 200)
})
data.head()
```

## Preprocessing

```python
scaler = StandardScaler()
scaled = scaler.fit_transform(data)
```

## KMeans Clustering

```python
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans_labels = kmeans.fit_predict(scaled)
```

## DBSCAN Clustering

```python
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(scaled)
```

## Visualization

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.scatterplot(x=scaled[:, 0], y=scaled[:, 1], hue=kmeans_labels, palette="tab10")
plt.title("KMeans Clustering")

plt.subplot(1, 2, 2)
sns.scatterplot(x=scaled[:, 0], y=scaled[:, 1], hue=dbscan_labels, palette="tab10")
plt.title("DBSCAN Clustering")
plt.tight_layout()
plt.show()
```
